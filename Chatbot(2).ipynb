```python
data={"intents":[
    {
        "tag":"greeting",
        "patterns":["hello","Hi there","Sup","Nice to meet you"],
        "responses":["Howdy partner","Hello","How are you doing"]
        
        
    },
    {
        "tag":"age",
        "patterns":["enter your age","DOB","Date of Birth","birthdate"],
        "responses":["22","17/06/2000","17th June 2000"]
        
        
    },
    {
        "tag":"meeting",
        "patterns":["there will be a meeting on this weekend ","meeting will be on Saturday at 6pm","morning meet"],
        "responses":["Yes Sir","What is the dress Code","Will it be online or offline at office "]
        
        
    },
    {
        "tag":"name",
        "patterns":["Enter name","what's your name ","who are you","enter full name "],
        "responses":["Nilaz D ROGER","Nilaz","I am NILAZ ROGER","Nilaz Donald ROGER"]
        
        
    },
    {
        "tag":"goodbye",
        "patterns":["bye",],
        "responses":["bye, it was a fruitful meeting "]
        
        
    }
]}
    
```

```python
import string 
import random
import json
import nltk
import numpy as np
from nltk.stem import WordNetLemmatizer

import tensorflow as tf
from keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout

nltk.download("punkt")
nltk.download("wordnet")
```

```python
lemmatizer=WordNetLemmatizer()
words=[]
classes=[]
doc_a=[]
doc_b=[]
```

```python
for intent in data["intents"]:
    for pattern in intent["patterns"]:
        tokens= nltk.word_tokenize(pattern)
        words.extend(tokens)
        doc_a.append(pattern)
        doc_b.append(intent["tag"])
        
    if intent["tag"] not in classes:
           classes.append(intent["tag"])
import nltk
nltk.download('omw-1.4')            
```

```python
words=[lemmatizer.lemmatize(word.lower()) for word in words if word not in string.punctuation]
```

```python
words=sorted(set(words))
classes=sorted(set(classes))
```

```python
training=[]
out_empty=[0]*len(classes)
for idx, doc in enumerate(doc_a):
    bow=[]
    text=lemmatizer.lemmatize(doc.lower())
    for word in words:
        bow.append(1) if word in text else bow.append(0)
    output_row= list(out_empty)
    output_row[classes.index(doc_b[idx])]=1
    training.append([bow, output_row])

random.shuffle(training)
training=np.array(training, dtype=object)

train_X=np.array(list(training[:,0]))
train_y=np.array(list(training[:,1]))
```

```python
input_shape=(len(train_X[0]),)
output_shape=len(train_y[0])
epochs=200
```

```python
    model=Sequential()
    model.add(Dense(128,input_shape=input_shape, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(64, activation="relu"))
    model.add(Dropout(0.3))
    model.add(Dense(output_shape, activation="softmax"))
    adam=tf.keras.optimizers.Adam(learning_rate=0.01, decay=1e-6)
   
    model.compile(loss="categorical_crossentropy",
                 optimizer=adam,
                 metrics=["accuracy"])
    
    print(model.summary())
    model.fit(x=train_X, y=train_y, epochs=200, verbose=1)
```

```python
def clean_text(text):
    tokens=nltk.word_tokenize(text)
    tokens=[lemmatizer.lemmatize(word) for word in tokens]
    return tokens
def bag_of_word(text,vocab):
    tokens=clean_text(text)
    bow=[0]*len(vocab)
    for w in tokens:
        for idx, word in enumerate(vocab):
            if word==w:
                bow[idx]=1
    return np.array(bow)
    
```

```python
def pred_class(text, vocab, labels):
    bow = bag_of_word(text,vocab)
    result= model.predict(np.array([bow]))[0]
    thresh=0.2
    y_pred=[[idx, res] for idx, res in enumerate(result) if res>thresh]
    
    y_pred.sort(key=lambda x:x[1], reverse=True)
    return_list=[]
    for r in y_pred:
        return_list.append(labels[r[0]])
        return return_list

def get_response(intents_list, intents_json):
    tag=intents_list[0]
    list_of_intents= intents_json["intents"]
    for i in list_of_intents:
        if i["tag"]==tag:
            result= random.choice(i["responses"])
            break
    return result
```
